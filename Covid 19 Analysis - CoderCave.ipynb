#### Comparing the US & India 

```python
df_us_india = df[df['Country/Region'].isin(['US', 'India'])][['Country/Region', 'Confirmed','Recovered']]
df_us_india = df_us_india.set_index('Country/Region')
df_us_india
```

#### Comparing the deaths/100 cases in different WHO Region

```python
sns.factorplot('Deaths / 100 Cases','WHO Region',data=df,
                   aspect=0.8,
                   height=15,
                   palette='magma',
                   join=False,
              )
```

We have the highest deaths/100 cases in the Eastern Mediterian Region and lowest in the Africa Region

#### Comapring the Recovered/100 Cases in different WHO Region

```python
sns.factorplot('Recovered / 100 Cases','WHO Region',data=df,
                   aspect=0.8,
                   height=15,
                   palette='magma',
                   join=False,
              )
```

The most of the recoveries happened in the South East Asia Region followed by Western Pacific

#### Comparing the Deaths , Recovered & Active Cases as per the different WHO Regions

```python
factor = ['Deaths', 'Recovered', 'Active']

for i in factor:
    byRegion=df.groupby('WHO Region').mean()
    byRegion.sort_values(by=[i], ascending=False, inplace=False)
    plt.figure(figsize=(25,10),dpi=100)
    sns.barplot(x = byRegion.index,y=byRegion[i],data = byRegion)
    plt.xticks(rotation=90)
    plt.xlabel('WHO Region')
    plt.ylabel(i)
    plt.title(i + 'According to WHO Region')
```

#### Top 10 with New Cases, New Recovered & New Deaths

```python
factor = ['New cases', 'New recovered', 'New deaths']

for i in factor:
    byRegion=df.groupby('Country/Region')[i].sum()
    Top_10=byRegion.sort_values(ascending=False, inplace=False).head(10)
    plt.figure(figsize=(25,10),dpi=100)
    sns.barplot(x = Top_10.index,y=Top_10.values)
    plt.xticks(rotation=90)
    plt.xlabel('Country/Region')
    plt.ylabel(i)
    plt.title('Top 10 Countries by {}'.format(i))
```
